To implement Simple Linear Regression using Python and evaluate its performance using MAE, MSE, and RMSE.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
# --------------------------------------------
# Step 1: Input Dataset
# --------------------------------------------

x = np.array([1, 2, 3, 4, 5])
y = np.array([20, 40, 50, 65, 80])

print("X =", x)
print("Y =", y)
```

```python
# Step 2: Calculate Mean
# --------------------------------------------

x_mean = np.mean(x)
y_mean = np.mean(y)

print("\nMean of X =", x_mean)
print("Mean of Y =", y_mean)
```

```python
# --------------------------------------------
# Step 3: Calculate Slope (β1)
# --------------------------------------------

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

slope = numerator / denominator

print("\nNumerator =", numerator)
print("Denominator =", denominator)
print("Slope (β1) =", slope)
```

```python
# --------------------------------------------
# Step 4: Calculate Intercept (β0)
# --------------------------------------------

intercept = y_mean - slope * x_mean

print("\nIntercept (β0) =", intercept)
```

```python
# --------------------------------------------
# Step 5: Regression Equation
# --------------------------------------------

print("\nRegression Equation")
print("y =", round(intercept, 2), "+", round(slope, 2), "* x")
```

```python
# --------------------------------------------
# Step 6: Prediction
# --------------------------------------------

y_pred = intercept + slope * x

print("\nPredicted Values")

for i in range(len(x)):
    print("x =", x[i], " Predicted =", round(y_pred[i], 2))
```

```python
# --------------------------------------------
# Step 7: Residual Errors
# --------------------------------------------

residual = y - y_pred

print("\nResidual Errors")

for i in range(len(x)):
    print("Actual =", y[i],
          " Predicted =", round(y_pred[i], 2),
          " Residual =", round(residual[i], 2))
```

```python
# --------------------------------------------
# Step 8: Performance Metrics
# --------------------------------------------

mae = mean_absolute_error(y, y_pred)

mse = mean_squared_error(y, y_pred)

rmse = np.sqrt(mse)

print("\nEvaluation Metrics")
print("MAE =", round(mae, 4))
print("MSE =", round(mse, 4))
print("RMSE =", round(rmse, 4))
```

```python
# --------------------------------------------
# Step 9: Scatter Plot with Regression Line
# --------------------------------------------

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Scatter plot (Actual Data)
plt.scatter(x, y,
            color='blue',
            s=80,
            label='Actual Data')
# Regression Line
plt.plot(x, y_pred,
         color='red',
         linewidth=2,
         label='Regression Line')

# Labels
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")

# Legend and Grid
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```
